```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv('train1.csv', index_col=0)
```

```python
# List of columns
train.columns
```

```python
# Data types of columns
train.dtypes
```

```python
train['sex'].value_counts()
```

```python
test = pd.read_csv("test.csv", index_col=0)
print(test.dtypes)
test.head()
```

```python
y = train['historical_popularity_index']
X = train[['page_views','average_views', 'article_languages']]
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = LinearRegression()
for i in range(4):
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    model.fit(X_train, y_train) # acutal training
    print(model.score(X_test, y_test))
```

```python
model.fit(X, y)
test['historical_popularity_index'] = model.predict(test[['page_views', 'average_views', 'article_languages']].values)
test.head()
```

```python
test[['historical_popularity_index']].to_csv('sample_submission.csv')
```

```python
train['birth_year'] = train['birth_year'].apply(pd.to_numeric, errors='coerce')
train['birth_year'] = train['birth_year'].fillna(0)
print(train.dtypes)
train.head()
```

```python
y = train['historical_popularity_index']
X = train[['page_views','average_views', 'article_languages', 'birth_year']]
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
test['historical_popularity_index'] = model.predict(test[['page_views','average_views', 'article_languages', 'birth_year']])
```

```python
test = test.drop(test.columns[0:-1], axis=1)
test.head()
```

```python
test.to_csv('slightly_better_submission.csv')
```
